<h2>5. Checking missing values in the dataset</h2>

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
```

```python
percentage_of_missing_values = ( data.isnull().sum() * 100 ) / len(data)
```

```python
percentage_of_missing_values
```

<p>From the above output, we can see that 12.8% values are missing in the revenue column and 6.4% values are missing in the metascore.</p>

<h2>6. Dropping all the missing values</h2>

```python
data.dropna(axis=0) # 'axis=0' means drop the rows which have missing values and 'axis=1' means drop columns which have missing values
```

<h2>7. Checking for duplicate data</h2>

```python
dup_data = data.duplicated().any()
```

```python
print("Are there any duplicate values in the dataset?",dup_data)
```

```python
# what if, we have duplicate data?
# in that case, we will follow the below approach to drop those duplicates

data = data.drop_duplicates()
data
```

```python
# for categorical as well as numerical, we need to follow the below approach

data.describe(include='all')
```

<h2>10. In which year, there was the highest average voting</h2>

```python
data.columns
```

```python
data.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

<h2>13. Displaying top 10 lengthy movies title and Runtime</h2>

```python
data.columns
```

```python
data.nlargest(10,'Runtime (Minutes)')
```

```python
data.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']]
```

```python
# another method

data.groupby('Title')['Runtime (Minutes)'].mean().sort_values(ascending=False).head(10)
```

```python
# another method

data.sort_values(['Runtime (Minutes)'], ascending=False).head(10)[['Title','Runtime (Minutes)']]
```

```python
# now let's set the index to title

top_10_lengthy_movies = data.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']].set_index('Title')
top_10_lengthy_movies
```

```python
sns.barplot(x='Runtime (Minutes)', y=top_10_lengthy_movies.index, data=top_10_lengthy_movies)
plt.title('Movies by Runtime')
plt.show()
```

<h2>16. Top 10 highest rated movie and their directors</h2>

```python
data.columns
```

```python
data.nlargest(n=10,columns='Rating')[['Title','Director','Rating']]
```

```python
top_10_movie_directors_rating = data.nlargest(n=10,columns='Rating')[['Title','Director','Rating']].set_index('Title')
top_10_movie_directors_rating
```

```python
sns.barplot(x='Rating',y=top_10_movie_directors_rating.index, data=top_10_movie_directors_rating)
plt.title('Movies by Rating')
plt.show()
```

```python
sns.barplot(x='Rating',y=top_10_movie_directors_rating.index, data=top_10_movie_directors_rating, hue='Director', dodge=False)
plt.title('Movie Title & Director by Rating')
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.show()
```

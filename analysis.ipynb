<h2>22. Finding unique values from Genre</h2>

```python
data.columns
```

```python
data['Genre']
```

<p>From above, output we can see that, each row contains multiple genres separated by comma.</p>

```python
# creating the list of all rows of data['Genre']

list_genre = []

for row in data['Genre']:
    list_genre.append(row)

list_genre
```

```python
# now, let's separate values taking delimiter as comma and update list_no_comma_genre

list_no_comma_genre = []

for item in list_genre:
    list_no_comma_genre.append(item.split(','))
    
list_no_comma_genre
```

```python
# creating unique genre list

unique_genre = []

for row in list_no_comma_genre:
    for genre in row:
        if genre not in unique_genre:
            unique_genre.append(genre)

unique_genre
```

```python
print('Total no of unique values in genre: ',len(unique_genre))
```

<h2>23. How many films of each genre were made?</h2>

```python
movies_with_genre = {}

for genre in unique_genre:
    genre_movie_count = 0
    for item in list_genre:
        if genre in item:
            genre_movie_count += 1
    movies_with_genre[genre] = genre_movie_count

movies_with_genre
```
